### Data loader
--- 
At this part, the following variables are defined: 
```
nii_data := list of path to the nifti images files.
patients := list of path to the directories with patient's pngs views. 
```

````python
data_path = '../testing_data/fpngs/'
path = '../testing_data/'
nii_data = [join(path, ni) for ni in os.listdir(path) if ni.endswith('.nii')]
patients = [join(data_path, patient) for patient in os.listdir(data_path) if os.path.isdir(join(data_path, patient))]
````

````python
selector = Select(options=patients)
selector
````

````python
imgs = [f for f in os.listdir(selector.value)]
imgs.sort(key = lambda x: x.split('-', maxsplit=1)[1][-6:-4])
imgs_path = [join(selector.value, img) for img in imgs]
len(imgs_path)
````

````python
img = Image.open(imgs_path[25])
img
````

````python
def print_img(count):
    img = Image.open(imgs_path[count])
    img
    return img

interact(print_img, count=widgets.IntSlider(min=0, max=len(imgs_path)-1, step=1, value=10))
````
